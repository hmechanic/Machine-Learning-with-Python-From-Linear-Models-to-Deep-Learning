## 10. Overlapping, multi-digit MNIST

### Fully connected network

#### Architecture Breakdown:

**Input Layer:**

Takes flattened image data (4228 = 1176 pixels)
The Flatten() layer converts the 2D image into a 1D vector


**Hidden Layers:**

- First hidden layer: 64 neurons with ReLU activation
- Second hidden layer: 64 neurons with ReLU activation
  
These layers learn shared representations useful for both digit recognition tasks


Output Layers:

Two separate output heads, each with 10 neurons (one for each digit 0-9)
- linear_first_digit: Classifies the first digit
- linear_second_digit: Classifies the second digit
  

No activation function (returns raw logits for CrossEntropy loss)



### Convolutional Model

** Feature Extraction**

The Sequential encoder block processes the input through:

1. Conv layers detect edges and patterns
2. ReLU adds non-linearity
3. MaxPool reduces spatial dimensions
4. Flatten prepares for classification


** Representation Learning**
LazyLinear learns shared representations:

- 128-dimensional feature vector
- Captures digit-relevant patterns
- Dropout prevents overfitting
- ReLU maintains non-linearity

** Multi-task Classification**
Separate heads for each digit:

- Shared features, specialized outputs
- 10 classes per digit (0-9)
- Simultaneous prediction
- Independent loss computation
