The learning goal og the player is to find a policy that $\pi: ~S \rightarrow C$ that maximizes the **expected cumulative dicounted reward**:
\begin{equation}
\mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^t R\left(h_t, a_t, b_t\right) \mid\left(a_t, b_t\right) \sim \pi\right]
\end{equation}

The optimal expected reward achievable is defined as:
\begin{equation}
V^*=\mathbb{E}_{h \sim \Gamma_0, s \sim \Psi(h)}\left[V^*(s)\right]
\end{equation}

Where:
\begin{equation}
V^*(s)=\max _\pi \mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^t R\left(h_t, a_t, b_t\right) \mid h_0=h(s), s_0=s,\left(a_t, b_t\right) \sim \pi\right] .
\end{equation}

We can define the optimal Q-function as:
\begin{equation}
Q^*(s, a, b)=\max _\pi \mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^t R\left(h_t, a_t, b_t\right) \mid h_0=h(s), s_0=s, a_0=a, b_0=b,\left(a_t, b_t\right) \sim \pi \text { for } t \geq 1\right] .
\end{equation}

Note that given $Q^* (s,a,b)$ we can obtain an optimal policy:
\begin{equation}
\pi^*(s)=\underset{(a, b) \in C}{\arg \max } Q^*(s, a, b) .
\end{equation}

## Relations between value function and Q-function
\begin{aligned}
\text{Step 1:} \quad & Q^*(s, a, b) = \max_\pi \mathbb{E} \left[ \sum_{t=0}^{\infty} \gamma^t R(h_t, a_t, b_t) \mid h_0 = h(s), s_0 = s, a_0 = a, b_0 = b \right] \\
\text{Step 2:} \quad & = \max_\pi \mathbb{E} \left[ R(s, a, b) + \sum_{t=1}^{\infty} \gamma^t R(h_t, a_t, b_t) \mid \cdots \right] \\
\text{Step 3:} \quad & = R(s, a, b) + \max_\pi \mathbb{E} \left[ \sum_{t=1}^{\infty} \gamma^t R(h_t, a_t, b_t) \mid \cdots \right] \\
\text{Step 4:} \quad & = R(s, a, b) + \gamma \max_\pi \mathbb{E} \left[ \sum_{t=1}^{\infty} \gamma^{t-1} R(h_t, a_t, b_t) \mid \cdots \right] \\
\text{Step 5:} \quad & = R(s, a, b) + \gamma \max_\pi \mathbb{E} \left[ \sum_{\tau=0}^{\infty} \gamma^{\tau} R(h_{\tau+1}, a_{\tau+1}, b_{\tau+1}) \mid \cdots \right] \\
\text{Step 6:} \quad & = R(s, a, b) + \gamma \, \mathbb{E} \left[ \max_\pi \mathbb{E} \left[ \sum_{\tau=0}^{\infty} \gamma^{\tau} R(h_{\tau}, a_{\tau}, b_{\tau}) \mid s_1 \right] \mid s, a, b \right] \\
\text{Step 7:} \quad & = R(s, a, b) + \gamma \, \mathbb{E} \left[ V^*(s_1) \mid s_0 = s, a_0 = a, b_0 = b \right]
\end{aligned}

## Optimal episodic reward

Let $V^*(h_0) $ be the optimal value function for an initial state $h_0$, i.e.,
\begin{equation}
V^*\left(h_0\right)=\mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^t R\left(h_t, a_t, b_t\right) \mid \pi^*\right]
\end{equation}
Compute the expected optimal reward for each episode $\mathbb{E}\left[ V^*(h_0) \right]$, Note that the initial state is uniformly distributed in the state spece $ H = (r, q): 0 \leq r \leq 3, 0 \leq q \leq 3$.
