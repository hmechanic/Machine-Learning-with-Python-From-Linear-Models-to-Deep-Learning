## kernel perceptron algorithm with softmax regression
Recall the kernel perceptron algorithm. The weights $\theta$ can be represented by a linear combination of features:

$$
\theta = \sum _{i=1}^{n} \alpha ^{(i)} y^{(i)} \phi (x^{(i)})
$$

In the softmax regression fomulation, we can also apply this representation of the weights:

$$
\theta _ j = \sum _{i=1}^{n} \alpha _{j}^{(i)} y^{(i)} \phi (x^{(i)}).
$$

Recall the softmax linear regression:

\begin{equation}
h(x) = \frac{1}{\sum _{j=1}^ k e^{\theta _ j \cdot x / \tau }} \begin{bmatrix}  e^{\theta _1 \cdot x / \tau } \\ e^{\theta _2 \cdot x / \tau } \\ \vdots \\ e^{\theta _ k \cdot x / \tau } \end{bmatrix}
\end{equation}

where $h(x)$ is the probability vector (over all classes $k$ classes) given the feature vector $x$ (dimension is $d$) and the parameter vector $theta$ (dimension $k$ by $d$). 
$\tau$ is a hyperparameter that controls the variance of the distribution.

Remember that the cost function we want to minimize in order to find $\theta$ is:

\begin{equation}
J(\theta ) = -\frac{1}{n}\Bigg[\sum _{i=1}^ n \sum _{j=1}^{k} [[y^{(i)} == j]] \log {\frac{e^{\theta _ j \cdot x^{(i)} / \tau }}{\sum _{l=1}^ k e^{\theta _ l \cdot x^{(i)} / \tau }}}\Bigg] + \frac{\lambda }{2}\sum _{i=1}^{k}\sum _{j=0}^{d-1} \theta _{ij}^2
\end{equation}

and the gradient is:
\begin{equation}
\nabla _{\Theta _ j} J(\theta ) = -\frac{1}{\tau n} \sum _{i = 1} ^{n} \bigg[ x^{(i)}\big([[y^{(i)} == j]] - p(y^{(i)} = j | x^{(i)}, \theta )\big)\bigg] + \lambda \theta _ j
\end{equation}


Aplying the transformation to the softmax equation for non linear:

\begin{equation}
h(x) = \frac{1}{\sum _{j=1}^ k e^{[\theta _ j \cdot \phi (x) / \tau ] - c}} \begin{bmatrix}  e^{[\theta _1 \cdot \phi (x) / \tau ] - c} \\ e^{[\theta _2 \cdot \phi (x) / \tau ] - c} \\ \vdots \\ e^{[\theta _ k \cdot \phi (x) / \tau ] - c} 
\end{bmatrix}
\end{equation}

Replacing the weigths representation:

\begin{equation}
h(x) = \frac{1}{\sum _{j=1}^ k e^{[\sum _{i=1}^{n} \alpha _{j}^{(i)} y^{(i)} \phi (x^{(i)}) \cdot \phi (x) / \tau ] - c}} \begin{bmatrix}  e^{[\sum _{i=1}^{n} \alpha _{1}^{(i)} y^{(i)} \phi (x^{(i)}) \cdot \phi (x) / \tau ] - c} \\ e^{[\sum _{i=1}^{n} \alpha _{2}^{(i)} y^{(i)} \phi (x^{(i)}) \cdot \phi (x) / \tau ] - c} \\ \vdots \\ e^{[\sum _{i=1}^{n} \alpha _{k}^{(i)} y^{(i)} \phi (x^{(i)}) \cdot \phi (x) / \tau ] - c} 
\end{bmatrix}
\end{equation}

We actually do not need the real mapping $\phi (x)$, but the inner product between two features after mapping: $\phi (x_i) \cdot \phi (x)$, where $x_i$ is a point in the training set and $x$ is the new data point for which we want to compute the probability. If we can create a kernel function $K(x,y) = \phi (x) \cdot \phi (y)$, for any two points $x$ and $y$, we can then kernelize our softmax regression algorithm.




```python
def polynomial_kernel(X, Y, c, p):
    """
        Compute the polynomial kernel between two matrices X and Y::
            K(x, y) = (<x, y> + c)^p
        for each pair of rows x in X and y in Y.

        Args:
            X - (n, d) NumPy array (n datapoints each with d features)
            Y - (m, d) NumPy array (m datapoints each with d features)
            c - a coefficient to trade off high-order and low-order terms (scalar)
            p - the degree of the polynomial kernel

        Returns:
            kernel_matrix - (n, m) Numpy array containing the kernel matrix
    """
    # 1. Compute the matrix of all pairwise dot products.
    #    The result is an (n, m) matrix.
    dot_products = X @ Y.T
    
    # 2. Add the scalar 'c' and raise to the power 'p'.
    #    NumPy handles these operations element-wise.
    kernel_matrix = (dot_products + c) ** p
    
    return kernel_matrix

def rbf_kernel(X, Y, gamma):
    """
        Compute the Gaussian RBF kernel between two matrices X and Y::
            K(x, y) = exp(-gamma ||x-y||^2)
        for each pair of rows x in X and y in Y.

        Args:
            X - (n, d) NumPy array (n datapoints each with d features)
            Y - (m, d) NumPy array (m datapoints each with d features)
            gamma - the gamma parameter of gaussian function (scalar)

        Returns:
            kernel_matrix - (n, m) Numpy array containing the kernel matrix
    """
    # YOUR CODE HERE
    
    Xnorm = np.sum(X**2, axis=1)
    Ynorm = np.sum(Y**2, axis=1)
    cross = -2. * X @ Y.T
    kernel_matrix = np.exp(-gamma *(Xnorm[:,None] + Ynorm[None,:] + cross))
    
    return kernel_matrix
    
```

```python
def compute_probabilities_kernelized(X_new, X_train, y_train, alpha, temp_parameter, kernel_func, **kernel_args):
    """
    Computes, for each datapoint X_new[i], the probability that it is labeled as j
    for j = 0, 1, ..., k-1, using a kernelized model.

    Args:
        X_new - (m, d) NumPy array (m new datapoints to classify)
        X_train - (n, d) NumPy array (n training datapoints)
        y_train - (n,) NumPy array (labels of the training data)
        alpha - (k, n) NumPy array, dual parameters of the model.
        temp_parameter - the temperature parameter of softmax function (scalar)
        kernel_func - a function that computes the kernel matrix.
        **kernel_args - additional arguments for the kernel function (e.g., c and p for polynomial_kernel)

    Returns:
        H - (k, m) NumPy array, where each entry H[j][i] is the probability 
            that X_new[i] is labeled as j.
    """
    n_train = X_train.shape[0]
    k_classes = alpha.shape[0]
    
    # 1. Compute the kernel matrix between training data and new data.
    #    The result is an (n, m) matrix.
    #    K[i, j] = K(X_train[i], X_new[j])
    kernel_matrix = kernel_func(X_train, X_new, **kernel_args)

    # 2. Compute the weighted alpha terms.
    #    alpha is (k, n) and y_train is (n,). We reshape y_train to (1, n)
    #    to ensure it broadcasts correctly, multiplying each column of alpha
    #    by the corresponding y_train value.
    #    This corresponds to alpha_{j}^{(i)} * y^{(i)} in the formula.
    weighted_alpha = alpha * y_train.reshape(1, n_train) # Result is (k, n)

    # 3. Compute the raw scores for each class on each new data point.
    #    This performs the sum: sum_{i=1 to n} [weighted_alpha * K]
    #    (k, n) @ (n, m) -> (k, m) matrix of scores.
    raw_scores = weighted_alpha @ kernel_matrix

    # 4. Apply the temperature parameter.
    scores = raw_scores / temp_parameter

    # 5. For stability, subtract the max score *for each data point* (column-wise).
    #    We find the max along axis=0 (across the k scores for each point)
    #    and use keepdims=True to ensure the result can be broadcast correctly.
    max_scores = np.max(scores, axis=0, keepdims=True) # Result is (1, m)
    stable_scores = scores - max_scores

    # 6. Exponentiate the stable scores.
    exp_scores = np.exp(stable_scores)

    # 7. Sum the exponentiated scores for each data point to get the denominators.
    #    We sum along axis=0 and use keepdims=True for division broadcasting.
    denominators = np.sum(exp_scores, axis=0, keepdims=True) # Result is (1, m)

    # 8. Compute the final probabilities. The result is already the desired (k, m) shape.
    H = exp_scores / denominators

    return H
```

the objective function is:

\begin{equation}
J(\alpha)=-\frac{1}{n}\left[ \sum_{i=1}^{n}\sum_{j=1}^k [[y^{(i)}==j]]\text{log}\frac{e^{\sum_{m=1}^{n}\alpha_j^{(m)} K(x^{(m)},x^{(i)})/\tau}}{\sum_{l=1}^{k}e^{\sum_{m=1}^{n}\alpha_l^{(m)} K(x^{(m)},x^{(i)})/\tau}} \right] + \frac{\lambda}{2}\sum_{j=1}^{k}Tr\left(\alpha \cdot K_{train}\cdot\alpha^T \right)
\end{equation}

Calcular Probabilidades: Llamamos a nuestra función para obtener la matriz de probabilidades H (de dimensiones k x n) para los datos de entrenamiento. Es crucial entender que para calcular el coste sobre el conjunto de entrenamiento, $X_{new}$ y $X_{train}$ son el mismo conjunto de datos.

```python
def compute_cost_kernelized(X_train, y_train, alpha, lambda_factor, temp_parameter, kernel_func, **kernel_args):
    """
    Calcula el coste total sobre el conjunto de datos de entrenamiento para el modelo kernelizado.

    Args:
        X_train - (n, d) NumPy array (n datapoints cada uno con d features)
        y_train - (n, ) NumPy array conteniendo las etiquetas (un número de 0 a k-1)
        alpha - (k, n) NumPy array, los parámetros duales de nuestro modelo.
        lambda_factor - la constante de regularización (escalar)
        temp_parameter - el parámetro de temperatura de la función softmax (escalar)
        kernel_func - una función que calcula la matriz del kernel.
        **kernel_args - argumentos adicionales para la función del kernel (e.g., c y p)

    Returns
        c - el valor del coste (escalar)
    """
    n, d = X_train.shape
    k, _ = alpha.shape

    # == Parte 1: Coste de Ajuste (Cross-Entropy Loss) ==
    
    # 1. Calcula las probabilidades para cada punto de entrenamiento.
    #    Llamamos a la función de probabilidad usando los datos de entrenamiento tanto
    #    para la "base de conocimiento" (X_train) como para los puntos a evaluar (X_new).
    #    El resultado H es una matriz (k, n).
    H = compute_probabilities_kernelized(X_train, X_train, y_train, alpha, 
                                         temp_parameter, kernel_func, **kernel_args)

    # 2. Selecciona las log-probabilidades correspondientes a las etiquetas verdaderas.
    #    Para cada columna i (punto de dato i), seleccionamos la fila y_train[i].
    #    np.arange(n) -> [0, 1, ..., n-1] (índices de columna)
    #    y_train      -> [y_0, y_1, ..., y_{n-1}] (índices de fila)
    correct_probs = H[y_train, np.arange(n)]
    
    # 3. Calcula el coste de ajuste (pérdida de entropía cruzada media).
    #    Añadimos un pequeño épsilon por estabilidad numérica en caso de que una probabilidad sea 0.
    epsilon = 1e-9
    fit_cost = -np.mean(np.log(correct_probs + epsilon))

    # == Parte 2: Coste de Regularización ==
    
    # 1. Calcula la matriz del kernel del conjunto de entrenamiento consigo mismo.
    #    Esta matriz K (n, n) es la base para el cálculo de la regularización.
    K_train = kernel_func(X_train, X_train, **kernel_args)
    
    # 2. Calcula el término de regularización dual.
    #    La fórmula es (lambda/2) * Tr(alpha @ K @ alpha.T).
    #    (k, n) @ (n, n) @ (n, k) -> (k, k)
    regularization_term = alpha @ K_train @ alpha.T
    
    # 3. Calcula el coste de regularización final.
    reg_cost = (lambda_factor / 2) * np.trace(regularization_term)
    
    # == Coste Final ==
    total_cost = fit_cost + reg_cost
    
    return total_cost
```

The gradient is then:

\begin{equation}
\nabla _{\alpha_j^{(m)}}J(\alpha)=-\frac{1}{n\tau}\left[ \sum_{i=1}^{n} \left( [[y^{(i)}==j]] - p(y^{(i)} = j|x^{(i)},\alpha_j) \right) K(x^{(m)},x^{(i)})  \right] +\lambda\alpha_j K
\end{equation}

```python
def run_kernelized_gradient_descent_iteration(X_train, y_train, alpha_params, 
                                                learning_rate, lambda_factor, temp_parameter, 
                                                kernel_func, **kernel_args):
    """
    Ejecuta un paso de descenso de gradiente por lotes para la versión kernelizada.

    Args:
        X_train - (n, d) NumPy array (puntos de datos de entrenamiento)
        y_train - (n, ) NumPy array con las etiquetas
        alpha_params - (k, n) NumPy array, los parámetros duales actuales de nuestro modelo.
        learning_rate - la tasa de aprendizaje (escalar)
        lambda_factor - la constante de regularización (escalar)
        temp_parameter - el parámetro de temperatura de la función softmax (escalar)
        kernel_func - la función que calcula la matriz del kernel.
        **kernel_args - argumentos adicionales para la función del kernel.

    Returns:
        new_alpha - (k, n) NumPy array que es el valor final de los parámetros alpha después de un paso.
    """
    n, d = X_train.shape
    k, _ = alpha_params.shape

    # --- Paso 1: Calcular la Matriz del Kernel ---
    # La necesitamos tanto para las probabilidades como para el gradiente.
    K_train = kernel_func(X_train, X_train, **kernel_args)

    # --- Paso 2: Calcular Probabilidades y Matriz de Error ---
    # H es (k, n)
    H = compute_probabilities_kernelized(X_train, X_train, y_train, alpha_params, 
                                         temp_parameter, kernel_func, **kernel_args)
    
    # Crear la matriz Y one-hot (Y_matrix) usando el mismo método eficiente
    data = np.ones(n)
    rows = y_train
    cols = np.arange(n)
    Y_matrix = sparse.coo_matrix((data, (rows, cols)), shape=(k, n)).toarray()

    # Matriz de error (k, n)
    error_matrix = H - Y_matrix

    # --- Paso 3: Calcular el Gradiente ---
    # Normalizamos el gradiente de la pérdida por n, pero no por temp_parameter, 
    # ya que temp_parameter ya está incluido en el cálculo de H.
    
    # Gradiente de la pérdida: (k, n) @ (n, n) -> (k, n)
    loss_gradient = (1 / n) * error_matrix @ K_train
    
    # Gradiente de la regularización: (k, n) @ (n, n) -> (k, n)
    reg_gradient = lambda_factor * alpha_params @ K_train

    # Gradiente total
    cost_gradient = loss_gradient + reg_gradient

    # --- Paso 4: Actualizar los Parámetros alpha ---
    new_alpha = alpha_params - learning_rate * cost_gradient
    
    return new_alpha
```

```python

k = 10
n, _ = train_x.shape

# --- Hiperparámetros del Entrenamiento ---
learning_rate = 0.1
lambda_reg = 0.001
temp_parameter = 1.0
epochs = 50

# --- Parámetros del Kernel ---
c = 1
p = 3
gamma = 0.8

# --- Bucle de Entrenamiento ---
# Inicializar los parámetros duales alpha
alpha_params = np.zeros((k, n))

for i in range(epochs):
    alpha_params = run_kernelized_gradient_descent_iteration(
        train_x, train_y, alpha_params,
        learning_rate, lambda_reg, temp_parameter,
        rbf_kernel, gamma=gamma
    )
    
    if (i + 1) % 10 == 0:
        # Opcional: Calcular y mostrar el coste para ver el progreso
        cost = compute_cost_kernelized(train_x, train_y, alpha_params, lambda_reg, temp_parameter, rbf_kernel,gamma=gamma)
        print(f"Época {i+1}/{epochs}, Coste: {cost:.4f}")

print("\nEntrenamiento completado.")
print("Dimensiones de los parámetros alpha finales:", alpha_params.shape)
```

```python
def predict(X_new, X_train, y_train, alpha_params, temp_parameter, kernel_func, **kernel_args):
    """
    Predice las etiquetas para un nuevo conjunto de datos usando el modelo kernelizado entrenado.

    Args:
        X_new - (m, d) NumPy array (m nuevos puntos a clasificar)
        X_train - (n, d) NumPy array (puntos de datos de entrenamiento)
        y_train - (n, ) NumPy array con las etiquetas de entrenamiento
        alpha_params - (k, n) NumPy array, los parámetros duales ENTRENADOS.
        temp_parameter - el parámetro de temperatura de la función softmax (escalar).
        kernel_func - la función que calcula la matriz del kernel.
        **kernel_args - argumentos adicionales para la función del kernel.

    Returns:
        predictions - (m, ) NumPy array conteniendo la etiqueta predicha para cada
                      punto en X_new.
    """
    
    # Paso 1: Calcular la matriz de probabilidades (k, m) para los nuevos datos.
    # Nota: Los hiperparámetros de entrenamiento como lambda_factor y learning_rate
    # NO son necesarios para la predicción.
    H = compute_probabilities_kernelized(X_new, X_train, y_train, alpha_params, 
                                         temp_parameter, kernel_func, **kernel_args)
    
    # Paso 2: Encontrar el índice de la máxima probabilidad para cada columna (punto de dato).
    # np.argmax a lo largo del eje 0 encuentra el índice de la fila (clase) con el
    # valor más alto para cada columna.
    predictions = np.argmax(H, axis=0)
    
    return predictions
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['figure.figsize'] = [4, 4]
n = 200 # Aumentado para un mejor ejemplo
d = 2 
x = np.random.rand(n, d)

# Frontera de decisión circular (datos no separables linealmente)
y = ((x - 0.5)**2 @ np.array([1, 1]) > 0.08).astype(np.int32)

# Visualizar datos iniciales
print("Visualizando los datos de entrenamiento generados...")
plt.scatter(x[:,0], x[:,1], c=y, cmap=matplotlib.colors.ListedColormap(["#FF6347", "#4682B4"]))
plt.title("Datos de Entrenamiento Generados")
plt.show()
```

```python
# --- 2. Entrenamiento del Modelo ---
k = 2 # Número de clases
n, _ = x.shape

# --- Hiperparámetros ---
learning_rate = 0.5
lambda_reg = 0.0001
temp_parameter = 0.5
epochs = 150
gamma = 50 # Gamma es muy sensible, este valor funciona bien para estos datos

# --- Bucle de Entrenamiento ---
print("Iniciando el entrenamiento del modelo kernelizado...")
alpha_params = np.zeros((k, n)) # Inicializar alpha

for i in range(epochs):
    alpha_params = run_kernelized_gradient_descent_iteration(
        x, y, alpha_params,
        learning_rate, lambda_reg, temp_parameter,
        rbf_kernel, gamma=gamma
    )

    if (i + 1) % 10 == 0:
        cost = compute_cost_kernelized(x, y, alpha_params, lambda_reg, temp_parameter, rbf_kernel, gamma=gamma)
        print(f"Época {i+1}/{epochs}, Coste: {cost:.4f}")

print("\nEntrenamiento completado.")
```

```python
# --- 3. Visualización de la Frontera de Decisión ---
print("Generando la visualización de la frontera de decisión...")

# Crear una malla de puntos para cubrir todo el espacio
xx, yy = np.meshgrid(np.arange(0, 1.01, 0.02), np.arange(0, 1.01, 0.02))
xtest = np.vstack((xx.reshape(-1), yy.reshape(-1))).T

# Predecir la clase para cada punto en la malla
yhat = predict(xtest, x, y, alpha_params, temp_parameter, rbf_kernel, gamma=gamma)

# *** CORRECCIÓN CLAVE 1: Redimensionar yhat para el gráfico ***
yhat_reshaped = yhat.reshape(xx.shape)

# Graficar el fondo de color (la frontera de decisión)
plt.contourf(xx, yy, yhat_reshaped, cmap=plt.get_cmap("RdBu"), alpha=0.5)

# Superponer los puntos de entrenamiento originales
plt.scatter(x[:,0], x[:,1], c=y, cmap=matplotlib.colors.ListedColormap(['#FF0000', "#0000FF"]))
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.title("Frontera de Decisión del Modelo Kernelizado")
plt.show()
```
